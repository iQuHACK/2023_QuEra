```python
import networkx as nx
```

```python

n = 2
a = nx.Graph()
a.clear()
a.add_nodes_from([i for i in range(n)])
#block_edges = [(0,0), (2, 2),(4, 2),(6,0), (2,4), (4, 4),(0,6),(6,6)]
block_edges = [(0,1),(0,3),(0,4),(1,2),(1,5),(2,3),(2,6),(3,7),(4,5),(4,7),(5,6),(6,7)]
#block_edges = [(0,0), (0, 2)]
pos = {0: (0,0), 1: (6, 0), 2: (6, 6), 3: (0, 6), 4: (2,2), 5: (4,2), 6: (4,4), 7: (2,4)}
a.add_edges_from(block_edges)
nx.draw(a, pos)
```
