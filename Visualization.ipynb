```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import time
```

```python
fig = plt.figure(figsize=(10,10))
ax = plt.axes(projection='3d')
ax.axes.set_zlim3d(bottom=0, top=15) 
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('V')
x = np.linspace(-1.5, 1.5)
y = np.linspace(-1.5, 1.5)
X, Y = np.meshgrid(x, y)

def atom(h,k):
    return np.divide(1,np.sqrt((X-h) ** 2 + (Y-k) ** 2))
    
atoms = [(-0.01,-0.69),
(-0.66,0.35),
(0.68,0.36),
(-0.01,-0.02)]
Z = 0
for (h,k) in atoms:
    Z += atom(h,k)
ax.plot_wireframe(X, Y, Z, color ='orange')
plt.axis('on')
```
